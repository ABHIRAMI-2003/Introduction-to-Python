```python
import matplotlib.pyplot as plt
```

```python
X=[5,10,15]
Y=[8,13,15]
```

```python
plt.bar(X,Y, width=1.0)
```

```python
plt.barh(X,Y)
```

```python
X1=[1,3,4,5,6,7,9]
Y1=[4,7,2,4,7,8,3]

X2=[2,4,6,8,10]
Y2=[5,6,2,6,2]

plt.bar(X1,Y1,label="blue bar",color='b')
plt.bar(X2,Y2,label="green bar",color='g')
plt.plot()

plt.xlabel("bar number")
plt.ylabel("bar height")
plt.title("BAR cHART TITLE")
plt.legend()
plt.show()
```
